```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('assets/cleaned_car_data.csv')
df
```

```python
df = df.drop(columns = ['msrp'])
df.head()
```

```python
X = df.drop(columns = ['price_segment'])
y = df['price_segment']

X_encoded = pd.get_dummies(X, drop_first=False)

le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)
```

```python
lrc = LogisticRegression().fit(X_train, y_train)
```

```python
# recall 
y_lrc_train_pred = lrc.predict(X_train)

y_lrc_test_pred = lrc.predict(X_test)
```

```python
lrc_accuracy_score = accuracy_score(y_test, y_lrc_test_pred)
lrc_accuracy_score
```

### Implement Decision Tree Classification

```python
tree = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
#recall
y_tree_train_pred = tree.predict(X_train)

y_tree_test_pred = tree.predict(X_test)
```

```python
tree_accuracy_score = accuracy_score(y_test, y_tree_test_pred)
tree_accuracy_score
```

### Implementing Random Forest Classification

```python
rfc = RandomForestClassifier().fit(X_train, y_train)
```

```python
y_rfc_train_pred = rfc.predict(X_train)

y_rfc_test_pred = rfc.predict(X_test)
```

```python
rfc_accuracy_score = accuracy_score(y_test, y_rfc_test_pred)
rfc_accuracy_score
```

### Implementing XGBoost Classifier

```python
svm = SVC(C = 1.0, kernel='linear').fit(X_train, y_train) 
```

```python
y_svm_train_pred = svm.predict(X_train)

y_svm_test_pred = svm.predict(X_test)
```

```python
svm_accuracy_score = accuracy_score(y_test, y_svm_test_pred)
svm_accuracy_score
```
