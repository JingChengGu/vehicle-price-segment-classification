```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('assets/cleaned_car_data.csv')
df
```

```python
df = df.drop(columns = ['msrp'])
df.head()
```

```python
X = df.drop(columns = ['price_segment'])
y = df['price_segment']

X_encoded = pd.get_dummies(X, drop_first=False)

le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)
```

### Implementing Logistics Regression

```python
lrc = LogisticRegression().fit(X_train, y_train)
```

```python
lrc_cv_score = cross_val_score(lrc, X_encoded, y_encoded, cv = 5)
lrc_cv_score
```

```python
lrc_cv_score.mean()
```

```python
# recall 
y_lrc_train_pred = lrc.predict(X_train)

y_lrc_test_pred = lrc.predict(X_test)
```

```python
lrc_accuracy_score = accuracy_score(y_test, y_lrc_test_pred)
lrc_accuracy_score
```

### Implement Decision Tree Classification

```python
tree = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
tree_cv_score = cross_val_score(tree, X_encoded, y_encoded, cv = 5)
tree_cv_score.mean()
```

```python
#recall
y_tree_train_pred = tree.predict(X_train)

y_tree_test_pred = tree.predict(X_test)
```

```python
tree_accuracy_score = accuracy_score(y_test, y_tree_test_pred)
tree_accuracy_score
```

### Implementing Random Forest Classification

```python
rfc = RandomForestClassifier().fit(X_train, y_train)
```

```python
rfc_cv_score = cross_val_score(rfc, X_encoded, y_encoded, cv = 5)
rfc_cv_score.mean()
```

```python
y_rfc_train_pred = rfc.predict(X_train)

y_rfc_test_pred = rfc.predict(X_test)
```

```python
rfc_accuracy_score = accuracy_score(y_test, y_rfc_test_pred)
rfc_accuracy_score
```

### Implementing KNN

```python
X_scaled = StandardScaler().fit_transform(X_encoded)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size = 0.2, random_state=42)
```

```python
# K-fold validation on KNN between 20 k numbers
k_values = range(1, 21) # create [1-20]
cv_scores = [] # empty array to store our cross validation scores to extract the highest k
for k in k_values:
    knn = KNeighborsClassifier(k).fit(X_train, y_train)
    score = cross_val_score(knn, X_train, y_train, cv = 5)
    cv_scores.append(score.mean())

k_best_choice = k_values[np.argmax(cv_scores)]
k_best_choice
```

```python
knn = KNeighborsClassifier(k_best_choice).fit(X_train, y_train)
```

```python
y_knn_test_pred = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, y_knn_test_pred)
knn_accuracy
```
